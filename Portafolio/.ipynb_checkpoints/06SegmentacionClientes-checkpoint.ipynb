## **Anlisis de Datos Segmentacin de clientes potenciales (Customer Segmentation)**


### **Comprension del negocio**

**INTRODUCCION**

La segmentacin de clientes es la prctica de dividir una base de clientes en grupos de individuos que son similares en aspectos especficos relevantes para el marketing, como la edad, el gnero, los intereses y los hbitos de gasto.

Las empresas que emplean la segmentacin de clientes parten del hecho de que cada cliente es diferente, en preferencias y necesidades, y de que los esfuerzos de marketing se veran ms favorecidos si se dirigen mensajes a grupos de segmentos especficos.

**ENTORNO DE LA EMPRESA**

Una empresa de consumo tiene previsto entrar en nuevos mercados con sus productos actuales (P1, P2, P3, P4 y P5). Tras un intenso estudio de mercado, han deducido que el comportamiento del nuevo mercado es similar al de su mercado actual.

En el mercado actual, el equipo de ventas ha clasificado a todos los clientes en 4 segmentos (A, B, C, D ). A continuacin, han llevado a cabo una difusin y comunicacin segmentada para un segmento diferente de clientes. Esta estrategia les ha funcionado excepcionalmente bien. Planean utilizar la misma estrategia para los clientes potenciales en los nuevos mercados.

**OBJETIVO**

En este proyecto de aprendizaje automtico, nuestro objetivo es crear un modelo para predecir adecuadamente la Categorizacion/Clasificacion de los nuevos clientes.

### **Estructura del dataset**

***El conjunto de datos contiene informacin de clientes en 11 variables, donde 9 variables son caracteristicas demograficas, que permite a la empresa clasificar en segmentos de clientes para fines de marketing.***


- **ID:** Identificacin nica

- **Gender:** Gnero del cliente

- **Ever_Married:** Estado civil del cliente

- **Age:** Edad del cliente

- **Graduated:** Es el cliente un graduado?

- **Profession:** Profesin del cliente

- **Work_Experience:** Experiencia laboral en aos

- **Spending_Score:** Puntuacin del gasto del cliente

- **Family_Size:** Nmero de miembros de la familia del cliente.

- **Var_1:** Categora annima del cliente

- **Segmentation:** (target) Segmento del cliente

#### **Ingeniera de Variables**

*En el conjunto de datos hay variables categricas por lo que, antes de entrenar el modelo, es necesario aplicar tnicas de one-hot-encoding...*

##### **Transformacin Gender-Ever_Married-Graduated**

Las siguientes variables pueden ser codificadas en la etiqueta: Gender Ever_Married Graduated Spending_Score.

```python
# Utilizamos Label encoding para las variables categoricas de dos valores (Si/No)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
columns = ['Gender','Ever_Married','Graduated']
for col in columns:
    df[col] = le.fit_transform(df[col])

df
```

##### **Transformacin Profession**
Variables Profesion a aplicar One Hot Encoding

-- Revisar la distribucion de los datos para ver si los datos estan desbalanceados o no.

```python
df['Profession'].value_counts()
```

La variable "Profesin" tiene 9 categoras. Pero slo las 3 iniciales son las de mayor frecuencia. Por lo tanto, el resto se agrupa en una nueva categora denominada "Other".

```python
df['Profession'] = df['Profession'].replace(['Lawyer','Executive','Marketing','Homemaker'],'Other')
```

```python
#df = pd.get_dummies(df, columns = ['Profession'])
df['Profession'] = df['Profession'].map({'Artist': 0, 'Healthcare':1, 'Entertainment':2, 'Engineer': 3, 'Doctor':4, 'Other':5})
```

##### **Transformacin Var_1**
Variables Var_1 a aplicar One Hot Encoding.

Revisar la distribucion de los datos para ver si los datos estan desbalanceados o no.

```python
df['Var_1'].value_counts()
```

La variable "Var_1" tiene 7 categoras. Pero slo las 3 iniciales son las de mayor frecuencia. Por lo tanto, el resto se agrupa en una nueva categora denominada "Other".

```python
df['Var_1'] = df['Var_1'].replace(['Cat_5','Cat_1','Cat_7','Cat_2'],'Other')
df['Var_1'].value_counts()
```

Existen dos variables categricas adicionales: Profesin y Var_1.
Como son variables nominales, hay que apicarles la tecnica One Hot Encoding (Variables Dummies)

```python
#df = pd.get_dummies(df, columns = ['Var_1'])
df['Var_1'] = df['Var_1'].map({'Cat_1': 0, 'Cat_2':1, 'Cat_3':2, 'Cat_4': 3, 'Cat_5':4, 'Cat_6':5, 'Cat_7':6})
```

##### **Transformacin 4-Spending_Score**
Variables Spending_Score a aplicar transformacin.

La variable 'Spending_Score' tiene tres valores : Low, Average and High. Como son variables ordinales, las codificare por separado.

```python
df['Spending_Score'] = df['Spending_Score'].map({'Low': 0, 'Average':1, 'High':2})
```

##### **Transformacin 5-Segmentation**
Variable Segmentation a aplicar transformacin.

La variable 'Segmentation' tiene los valores : A, B, C, D. Como son variables ordinales, las codificare por separado.

```python
df['Segmentation'] = df['Segmentation'].map({'A':0,'B':1,'C':2,'D':3})
```

## **Configurar setup del modelo**

Ejecutamos setup de Pycaret, proponiendo como variable dependiente *Segmentation*. Dejaremos que Pycaret transforme las variables categoricas con transformacion Dummies o Onehot Encoding

```python
cat_features = ['Gender', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score', 'Var_1']
```
