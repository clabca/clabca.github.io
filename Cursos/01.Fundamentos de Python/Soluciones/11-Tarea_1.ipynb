## <font color='green'>Actividad 9: Super challenging</font>
¿Qué tan bien comprendes la indexación (en listas) y el uso de las llaves (en diccionarios)?

Aquí te presentamos un conjunto de diccionarios y listas que están anidados (incluidos) dentro de un único diccionario $d$:

````python
# Ejecuta esta celda

d = {"nivel 1":[1, 2, {'nivel 2':[5, 6, [1,['sácame de aquí por favor !!!']]]}]}
````

Tu tarea es recuperar la cadena "sácame de aquí por favor !!!" del diccionario utilizando *indexing* y *keys*

Tip:
1. Aproxímate a la solución paso a paso, agregando lentamente más y más llamadas de índice.
2. Extraiga valores del diccionario usando **[name_of_key]**
3. Extraiga elementos de la lista usando indexing con **[número]**
4. Utiliza tantas celdas como necesites

````python
# Tu código aquí ...
# Buscamos la llave 'nivel 1' en el diccionario 'd'
d['nivel 1']
# Obtenemos una lista
````

````python
# Seleccionamos el tercer elemento de la lista (index 2, recordar que el índice parte en 0)
d['nivel 1'][2]
# Obtenemos otro diccionario
````

````python
# Seleccioanmos la llave 'nivel 2' en el diccionario
d['nivel 1'][2]['nivel 2']
# Obtenemos otra lista
````

````python
# De la lista extraemos el tercer elemento con index 2
d['nivel 1'][2]['nivel 2'][2]
# Obtenemos nuevamente una lista
````

````python
# Entonces seleccionamos el segundo elemento en la lista con index 1
d['nivel 1'][2]['nivel 2'][2][1]
# Atención !!! Obtenemos otra lista, NO el string que andamos buscando.
# Necesitamos un último paso.
````

````python
# Por último seleccionamos el primer elemento de la lista con index 0
d['nivel 1'][2]['nivel 2'][2][1][0]
# Bingo!! Ahora sí tenemos nuestro string
````

Tu salida debería verse como esto:

```Python
'sácame de aquí por favor !!!'
```
