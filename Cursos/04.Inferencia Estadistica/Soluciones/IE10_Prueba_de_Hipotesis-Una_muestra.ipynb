## <font color='blue'>**Definiciones**</font>

Una hip√≥tesis estad√≠stica es una afirmaci√≥n o proposici√≥n respecto a alguna caracter√≠stica de una poblaci√≥n, generalmente fundamentada sobre un par√°metro de la misma. Contrastar una hip√≥tesis es comparar las predicciones con la realidad que observamos ocurrida en una muestra. Si dentro del margen de error que estamos dispuestos a admitir, hay coincidencia, aceptaremos la hip√≥tesis y en caso contrario la rechazaremos.

1. La hip√≥tesis emitida se suele designar por¬†$ùëØ_ùüé$¬†y se llama¬†**Hip√≥tesis nula**. Lo de ‚Äúnula‚Äù viene de que partimos del supuesto que las diferencias entre el valor verdadero del par√°metro y su valor hipot√©tico, en realidad, no son tales sino debidas al azar, es decir, no hay diferencia entre ambos, o dicho de otra forma la diferencia es nula.

2. La hip√≥tesis contraria se designa por¬†$ùëØ_ùüè$¬†¬†y se llama¬†**Hip√≥tesis alternativa**.

### <font color='blue'>**Hip√≥tesis nula**</font>


La hip√≥tesis nula (denotada por $H_0$) es la afirmaci√≥n de que el valor de un par√°metro de la poblaci√≥n (media, proporci√≥n o desviaci√≥n est√°ndar) es igual a un valor aseverado.

Las siguientes son hip√≥tesis nulas t√≠picas:
<br>
<br>
1. $H_0: \quad \mu = 37$
2. $H_0: \quad p = 0.5$
2. $H_0: \quad \sigma = 15$
<br>
<br>

La hip√≥tesis nula se aprueba en forma directa, en el sentido que __se supone que es verdadera__, y llegamos a una conclusi√≥n de rechazar $H_0$ o no rechazar $H_0$.



### <font color='blue'>**Hip√≥tesis alternativa**</font>

La hip√≥tesis alternativa (denotada por $H_1$) es la afirmaci√≥n de que el par√°metro tiene un valor que difiere de la hip√≥tesis nula. Los s√≠mbolos que emplea la hip√≥tesis alternativa son: $\lt$, $\gt$ o bien $\ne$.
<br>
<br>
$$
\begin{align*}
\text{Medias}: &\quad H_1:\mu \gt 37 \quad H_1: \mu \lt 37 &\quad H_1: \mu \neq 37\\
\text{Proporciones}: &\quad H_1:p \gt 0.5 \quad H_1: p \lt 0.5 &\quad H_1: p \neq 0.5 \\
\text{Desviaci√≥n est√°ndar}: &\quad \underbrace{H_1:\sigma \gt 15 \quad H_1: \sigma \lt 15 }_{Unilaterales\\(una \, cola)}&\quad \underbrace{H_1: \sigma \neq 15}_{Bilaterales\\(dos \, colas)}
\end{align*}
$$
<br>
<br>
Si est√° realizando un estudio y desea emplear una prueba de hip√≥tesis para sustentar una aseveraci√≥n, esta debe redactarse de tal manera que se convierta en la hip√≥tesis alternativa. Esto quiere decir, que su aseveraci√≥n debe expresarse utilizando solo estos s√≠mbolos $\lt$, $\gt$ o bien $\ne$.

### <font color='blue'>**Ejemplos**</font>

1. Sospechamos que las bolsas de frutos secos de 100 gramos, realmente no pesan 100 gramos. Para contrastar esta hip√≥tesis plantear√≠amos:

$$H_0: \mu = 100$$
$$H_1: \mu \neq 100$$

2. Pensamos que la proporci√≥n de gente que vot√≥ al partido A en las elecciones (35%) ahora es inferior ya que no lo han hecho muy bien. Para contrastar esta hip√≥tesis:

$$H_0:p \geq 0.35 $$
$$H_1:p < 0.35 $$

3. Estar√≠a contento de comprobar que no pueden demostrar que mi media de notas ha bajado de 6.2 como parecen indicar los √∫ltimos ex√°menes. Para contrastar esta hip√≥tesis:

$$H_0:\mu \ge 6.2 $$
$$H_1:\mu \lt 6.2 $$
<br>
### __Contraste de hip√≥tesis__
Se formula la hip√≥tesis sobre la poblaci√≥n (e.g., $\mu$, $p$, $\sigma$, etc.).  Entonces las conclusiones sobre la validez de la hip√≥tesis se basar√°n en la informaci√≥n de una muestra (e.g.,. $\bar{x}$, $\hat{p}$, $S$, etc).

La realizaci√≥n de un contraste de hip√≥tesis __no consiste en__ dicidir cu√°l de las dos hip√≥tesis ($H_0$ o $H_1$) es m√°s creible, sino en decidir si la muestra proporciona o no suficiente evidencia para descartar $H_0$. Mientras no tengamos suficiente evidencia para aceptar $H_1$ tendremos que creernos que $H_0$ es cierta.

## <font color='blue'>**Estad√≠stico de prueba**</font>
El **estad√≠stico de prueba**, es un valor que se utiliza para tomar la decisi√≥n sobre la hip√≥tesis nula y se calcula convirtiendo el **estad√≠stico muestral** (tales como la media muestral $\bar{x}$, o la proporci√≥n muestral $\hat{p}$) en una puntuaci√≥n (tales como los $Z$ o $t$ _scores_), bajo el supuesto que la hip√≥tesis nula es verdadera. Este tipo de tests, utilizar√° puntuaciones $Z$ o $t$ dependiendo del tama√±o de la muestra y de si $\sigma$, (desviaci√≥n est√°ndar poblacional) es conocida.

### <font color='blue'>**Prueba $Z$**</font>

La __prueba $Z$__, es un tipo de prueba param√©trica que se utiliza para determinar el grado de significatividad estad√≠stica de las diferencias entre las medias de dos conjuntos de datos. Se utiliza cuando las muestras son amplias (de m√°s de 30 individuos aproximadamente), independientes (esto es, no correlacionadas) y con $\sigma$ conocido. Opera con la media, la desviaci√≥n t√≠pica y el n√∫mero de individuos de cada muestra. Una vez calculado el valor de $Z$, se comparar√° con el valor cr√≠tico de $Z$ obtenido de una curva de distribuci√≥n normal. La hip√≥tesis nula sostiene que la diferencia entre medias no es significativa y ser√° rechazada si el valor obtenido en la prueba Z sobrepasa el valor cr√≠tico.

Para determinar si se acepta o rechaza la hip√≥tesis $H_0$, se define una valor de $Z$ cr√≠tico. Este valor puede ser sobre una cola o ambas colas como muestra la figura siguiente.

### <font color='blue'>**Valor-P (P-Value)**</font>
El __nivel cr√≠tio $P$__ (o __Valor-P__), es el nivel de significaci√≥n __m√°s peque√±o__ para el que la muestra obtenida obligar√≠a a rechazar la hip√≥tesis nula. En t√©rminos simples, el __Valor-P__ nos ayuda a diferenciar resultados que son producto del azar del muestreo, de resultados que son estad√≠sticamente significativos.
* El Valor-p es la probabilidad (calculada bajo $H_0$) de obtener un resultado que sea menos compatible con $H_0$ que el resultado obtenido con la muestra.
* El Valor-P depende de la muestra.
* Puede considerarse como el apoyo que la hip√≥tesis nula recibe de las observaciones de la muestra:
 * Si el Valor-P e __menor__ que el nivel de significaci√≥n ($\alpha$) prefijado, podemos decir que tenemos un resultado estad√≠sticamente significativo y que el apoyo a $H_0$ es escaso y, por tanto, $H_0$ debe rechazarse.
 * Si el Valor-P es __mayor o igual__ que el nivel de significaci√≥n ($\alpha$) prefijado, el apoyo a $H_0$ es suficiente y, por tanto, $H_0$ __NO__ debe rechazarse.
* El Valor-P, al ser una probabilidad, oscila entre 0 y 1.

### <font color='blue'>**Nivel de significaci√≥n ($\alpha$)**</font>

Los niveles de significaci√≥n son el complemento al intervalo de confianza de una distribuci√≥n y se utilizan para contrastar la hip√≥tesis nula ($H_0$) en una prueba de inferencia estad√≠stica. En otras palabras, los niveles de significaci√≥n son las probabilidades que dejamos fuera del intervalo de confianza de una distribuci√≥n y nos ayudan a determinar si el estad√≠stico de contraste est√° en zona de rechazo o no.

Los valores t√≠picos utilizados para un nivel de significaci√≥n son $0.05$ y $0.01$. Ronald Fisher dec√≠a:

> ‚ÄúUn fen√≥meno es demostrable experimentalmente cuando se conoce como conducir un experimento que raramente falla para darnos un resultado estad√≠sticamente significativo‚Äù.

De acuerdo con esta l√≥gica se entiende que, con un nivel de significaci√≥n del $5\%$, en promedio cinco de cada 100 veces que la hip√≥tesis nula sea cierta se rechazar√° por azar. Es decir, 5 de cada 100 veces que rechazamos la hip√≥tesis nos estaremos equivocando, pues se asume que es cierta y la prueba no demuestra ni su falsedad ni su certeza.

<br>



### <font color='blue'>**Criterios de decisi√≥n**</font>

El procedimiento convencional de prueba de hip√≥tesis requiere que __siempre probemos la hip√≥tesis nula__, de manera que nuestra conclusi√≥n inicial siempre ser√° una de las siguientes:

1. Rechazo la hip√≥tesis nula.
2. No rechazo la hip√≥tesis nula.

La decisi√≥n de rechazar o no una hip√≥tesis nula, se puede realizar a trav√©s de dos m√©todos; el __M√©todo cl√°sico__ o por __Valor P__:

__M√©todo cl√°sico__
<br>
* __Rechazo__ $H_0$  si el estad√≠stico de prueba ($Z$ o $t$) cae dentro de la Regi√≥n cr√≠tica.

* __No Rechazo__ $H_0$  si el estad√≠stico de prueba __NO__ cae dentro de la Regi√≥n cr√≠tica.

__Valor P__
<br>
* __Rechazo__ $H_0$  si el valor $P \le \alpha$, donde $\alpha$ es el nivel de significancia.
* __No Rechazo__ $H_0$  si el valor $P \gt \alpha$
<br><br>

## Paso 2: Definir hip√≥tesis

Hip√≥tesis nula (H0): Œº=180

## Paso 3: Definir tipo de prueba a aplicar

Dado que se busca contrastar si la media poblacional es distinta de 180, se utilizar√° una prueba de dos colas.

## Paso 4: Calcular Z cr√≠tico

Para un nivel de significancia del 5%, el valor cr√≠tico de Z se obtiene a partir de la distribuci√≥n normal est√°ndar. ## Paso 6: Definir regla de decisi√≥n

Si el valor absoluto del estad√≠stico de prueba ‚à£Z‚à£ es mayor que el valor cr√≠tico de Z (¬±1.96), rechazamos la hip√≥tesis nula. En caso contrario, no tenemos suficiente evidencia para rechazar la hip√≥tesis nula.

## Paso 7: Calcular valor-p

El valor-p se calcula utilizando la distribuci√≥n normal est√°ndar y se refiere a la probabilidad de obtener un estad√≠stico de prueba al menos tan extremo como el observado, bajo la suposici√≥n de que la hip√≥tesis nula es cierta.

## Paso 8: Interpretar y concluir

Si el valor-p es menor que el nivel de significancia (0.05 en este caso), rechazamos la hip√≥tesis nula en favor de la hip√≥tesis alternativa.

```
# Tu c√≥digo aqu√≠ ...

# Paso 1: Calcular estad√≠sticos de muestra
muestras = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
n = len(muestras)
media_muestral = sum(muestras) / n

# Paso 2: Definir hip√≥tesis
mu = 180

# Paso 3: Definir tipo de prueba a aplicar
# Es una prueba de dos colas, por lo que utilizaremos el valor cr√≠tico Z de ¬±1.96

# Paso 4: Calcular Z cr√≠tico
z_critico = 1.96

# Paso 5: Calcular estad√≠stico de prueba
sigma = 4
z = (media_muestral - mu) / (sigma / (n ** 0.5))

# Paso 6: Definir regla de decisi√≥n
if abs(z) > z_critico:
    decision = "Rechazar H0"
else:
    decision = "No rechazar H0"

# Paso 7: Calcular valor-p
import scipy.stats as stats

valor_p = 2 * (1 - stats.norm.cdf(abs(z)))

# Paso 8: Interpretar y concluir
nivel_significancia = 0.05

if valor_p < nivel_significancia:
    conclusion = "Se rechaza la hip√≥tesis nula en favor de la alternativa."
else:
    conclusion = "No hay suficiente evidencia para rechazar la hip√≥tesis nula."

# Imprimir resultados
print(f"Media muestral: {media_muestral}")
print(f"Estad√≠stico de prueba (Z): {z}")
print(f"Valor-p: {valor_p}")
print(f"Decisi√≥n: {decision}")
print(f"Conclusi√≥n: {conclusion}")
```

### <font color='blue'>**T-Test**</font>


__Test-T__ es una prueba en la que el estad√≠stico utilizado tiene una distribuci√≥n __t de Student__ si la hip√≥tesis nula es cierta. Se aplica cuando la poblaci√≥n estudiada sigue una distribuci√≥n normal, pero el tama√±o muestral es demasiado peque√±o como para que el estad√≠stico en el que est√° basada la inferencia est√© normalmente distribuido, utiliz√°ndose una estimaci√≥n de la desviaci√≥n t√≠pica en lugar del valor real.

Para calcular __t__ se utiliza la siguiente formula:

$$ t = \frac{\bar{x} - \mu}{\frac{S}{\sqrt{n}}}$$

Donde $S$ es la la desviaci√≥n est√°ndar muestral.

### Rentabilidad fondos de inversi√≥n

Sea X la variable ‚Äùrentabilidad de cierto tipo de fondos de inversi√≥n tras una apreciaci√≥n fuerte del Euro con respecto al D√≥lar‚Äù. Se considera que la media de esta variable es 15. Un economista afirma que dicha rentabilidad media ha variado, por lo que lleva a cabo un estudio en las condiciones rese√±adas anteriormente, sobre una muestra de 9 fondos cuya media muestral resulta ser de 15.308 y cuya varianza muestral es 0.193.

1. Especificando las hip√≥tesis necesarias, contrastar la afirmaci√≥n del economista al 5%.

2. A partir del resultado de punto 1, razonar si el intervalo de confianza para la media al 95% contendr√° o no al valor 15.

```
import numpy as np

# 1.  calcular estadisticos muestra
xbar = 15.308
S2 = 0.193 # varianza muestral
S = np.sqrt(S2) # desv. std muestral insesgada (asumimos)
n = 9

print(f'Xbarra:         {xbar:4.2f}')
print(f'S:              {S:4.2f}')
print(f'Tama√±o muestra: {n}')


# 2. Definir hip√≥tesis a contrastar (prueba bilateral - 2 colas)
#    H0:   Mu = 15
#    H1:   Mu ‚â† 15

# 3. como n < 30 y no se conoce Sigma, aplicaremos T Student
alpha = 0.05
mu = 15

# 4. calcular valor cr√≠tico de T
#    n-1 grados de libertad, donde n corresponde al n√∫mero de la muestra y 2 colas
Tcritico = stats.t.ppf(1-(alpha/2), df=(n-1))
print(f'T cr√≠tico: {Tcritico:4.2f}')


# 5. calcular Estadistico de prueba t
t =  (xbar - mu) / (S / np.sqrt(n))
print(f't: {t:4.2f}')

# 6. Define Regla de decision
#    no rechaza H0 si: -Tcritico <= t <= Tcritico
#    rechazar H0 si:  t < -Tcritico  o  t > Tcritico


# 7. calcular valor-p
pvalue = 1 - (stats.t.cdf(t, df=(n - 1)))
print(f'Valor-p: {pvalue*2:4.2f}') #2 colas
```

```
# Tu c√≥digo aqu√≠ ...

import scipy.stats as stats

# Paso 2: Definir la hip√≥tesis
O = 152
E = 14

# Paso 3: Identificar el estad√≠stico (chi-cuadrado)
chi_cuadrado = ((O - E)**2) / E

# Paso 5: Calcular el valor cr√≠tico
valor_critico = stats.chi2.ppf(0.95, df=1)

# Paso 6: Interpretar y concluir
if chi_cuadrado > valor_critico:
    decision = "Rechazar H0"
else:
    decision = "No rechazar H0"

# Paso 7: Calcular el Valor-P
valor_p = 1 - stats.chi2.cdf(chi_cuadrado, df=1)

# Imprimir resultados
print(f"Estad√≠stico de chi-cuadrado: {chi_cuadrado}")
print(f"Valor cr√≠tico: {valor_critico}")
print(f"Decisi√≥n: {decision}")
print(f"Valor-p: {valor_p}")
```

```
import numpy as np
import matplotlib.pyplot as plt

# Definir la distribuci√≥n de chi-cuadrado con 1 grado de libertad
x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df=1)

# Crear el gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuci√≥n Chi-cuadrado (1 grado de libertad)', color='blue')

# √Årea cr√≠tica
x_critico = np.linspace(valor_critico, 10, 1000)
y_critico = stats.chi2.pdf(x_critico, df=1)
plt.fill_between(x_critico, y_critico, color='red', alpha=0.5, label='√Årea Cr√≠tica (5%)')

# L√≠nea vertical para el estad√≠stico de prueba
#plt.axvline(chi_cuadrado, color='green', linestyle='dashed', linewidth=1, label='Estad√≠stico de Prueba')

# Etiquetas y leyenda
plt.xlabel('Valor de chi-cuadrado')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci√≥n Chi-cuadrado y Contraste de Hip√≥tesis')
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

El valor-p es extremadamente cercano a cero, lo que indica una evidencia muy fuerte en contra de la hip√≥tesis nula. ## <font color='blue'>**Errores Tipo I y II**</font>

Cuando probamos la  hip√≥tesis llegamos a la conclusi√≥n de rechazarla o no rechazarla. Tales conclusiones pueden ser correctas o incorrectas, incluso cuando se hace todo correctamente.

Ninguna prueba de hip√≥tesis es 100% cierta. Puesto que la prueba se basa en probabilidades, siempre existe la posibilidad de llegar a una conclusi√≥n incorrecta. Cuando se realiza una prueba de hip√≥tesis, se puede cometer dos tipos de error: __tipo I__ y __tipo II__. Los riesgos de estos dos errores est√°n inversamente relacionados y se determinan seg√∫n el nivel de significancia y la potencia de la prueba. Por lo tanto, se debe determinar qu√© error tiene consecuencias m√°s graves para su situaci√≥n antes de definir los riesgos.

<br>


### <font color='blue'>**Error de tipo I**</font>

Si se rechaza la hip√≥tesis nula __$H_0$__ cuando es verdadera, se comete un __error de tipo I__. La probabilidad de cometer un __error de tipo I__ es __$\alpha$__, que es el __nivel de significancia__, el cual es establecido a priori por el investigador para la prueba de hip√≥tesis en particular. Un $\alpha = 0.05$ indica que se est√° dispuesto a aceptar una probabilidad de 5% de estar equivocado al rechazar la hip√≥tesis nula. Para reducir este riesgo, se debe utilizar un valor menor para $\alpha$. Sin embargo, usar un valor menor para alfa significa que usted tendr√° menos probabilidad de detectar una diferencia si esta realmente existe.
<br><br>
$$\text{Nivel de significancia} = \alpha$$

<br>

### <font color='blue'>**Error de tipo II**</font>

Cuando la hip√≥tesis nula __$H_0$__ es falsa y no se la rechaza, se comete un __error de tipo II__. La probabilidad de cometer un error de __tipo II__ es __$\beta$__, la cual depende de la potencia de la prueba. Se puede reducir el riesgo de cometer un error de tipo II al asegurar de que la prueba tenga suficiente potencia. Para ello, debemos asegurarnos de que el tama√±o de la muestra sea lo suficientemente grande como para detectar una diferencia pr√°ctica cuando esta realmente exista.

$\beta$ depende principalmente del tama√±o de la muestra y de la magnitud del efecto; as√≠ es que si su t√≥pico es dif√≠cil de probar debido a lo dif√≠cil que es obtener muestras, o el efecto que busca es casi insignificante, es m√°s probable comoter este tipo de error.

La probabilidad de rechazar la hip√≥tesis nula cuando es falsa es igual a $1 - \beta$. Este es el objetivo del investigador.  Este valor es conocido como la __potencia de la prueba__.
<br><br>
$$\text{Potencia de la prueba} = 1 - \beta$$
<br><br>



### <font color='blue'>**Matriz de confusi√≥n**</font>

|   Decisi√≥n basada en la muestra   | $H_0$ es verdadera | $H_0$ es falsa |
|----------------------------------------|:------------------:|:--------------:|
| **No rechazar $H_0$**  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\alpha$)     |<font color='red'>__Error tipo II__ </font> <br> no rechazar $H_0$ cuando es falsa <br>(probabilidad = $\beta$) |
| **Rechazar $H_0$**   | <font color='red'>__Error tipo I__</font>  <br>rechazar $H_0$ cuando es verdadera <br>(probabilidad = $\alpha$)  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\beta$)|


### Ejemplo de error de tipo I y tipo II
Para entender la interrelaci√≥n entre los errores de __tipo I__ y __tipo II__, y para determinar cu√°l error tiene consecuencias m√°s graves para una situaci√≥n espec√≠fica, consideremos el siguiente ejemplo.

Un investigador m√©dico desea comparar la efectividad de dos medicamentos. Las hip√≥tesis nula y alternativa son:

>Hip√≥tesis nula ($H_0$): $\mu_1 = \mu_2$, los dos medicamentos tienen la misma eficacia.
>
>Hip√≥tesis alternativa ($H_1$): $\mu_1 \ne \mu_2$, los dos medicamentos no tienen la misma eficacia.

Un __error de tipo I__ se produce si el investigador rechaza la hip√≥tesis nula y concluye que los dos medicamentos son diferentes cuando, en realidad, no lo son. Si los medicamentos tienen la misma eficacia, el investigador podr√≠a considerar que este error no es muy grave, porque de todos modos los pacientes se beneficiar√≠an con el mismo nivel de eficacia independientemente del medicamento que tomen. Sin embargo, si se produce un __error de tipo II__, el investigador no rechaza la hip√≥tesis nula cuando debe rechazarla. Es decir, el investigador concluye que los medicamentos son iguales cuando en realidad son diferentes. Este error puede poner en riesgo la vida de los pacientes si se pone en venta el medicamento menos efectivo en lugar del medicamento m√°s efectivo.

Cuando realice las pruebas de hip√≥tesis, considere los riesgos de cometer errores de tipo I y tipo II. Si las consecuencias de cometer un tipo de error son m√°s graves o costosas que cometer el otro tipo de error, entonces elija un nivel de significancia ($\alpha$) y una potencia para la prueba ($1 - \beta$) que reflejen la gravedad relativa de esas consecuencias.

### Reflexi√≥n sobre los Errores Tipo I y II

Ambos errores (tipo I y tipo II) destruyen valor a los negocios, la buena noticia es que con las pruebas de hip√≥tesis podemos minimizarlos tomando mejores decisiones, impactando as√≠ el aprendizaje del equipo, el bienestar de los clientes y el crecimiento organizacional.
