### Creando los datasets y ajustando sus distribuciones.

Crear 3 datasets con respuestas de clientes a una encuesta de satisfacción de producto. Las encuestas tienen una escala de 1 a 7 (discreta), donde nota 1 es muy baja satisfacción y la nota 7 es muy alta satisfacción con el producto.
Los datasets son los siguientes:

1. **Capel**: crear un conjunto con 800 encuestas de satisfacción, que representan la historia del último trimestre de Capel. La satisfacción en promedio fue de **4.50** puntos y tuvo una distribición similar a la de la siguiente figura para que la use de referencia.

2. **Competencia**: crear un dataset con 500 encuestas de satisfacción, que representan la historia del último trimestre de la principal marca de la competencia. 3. **Le Pac**: crear un dataset con 320 encuestas de satisfacción, que representan la satisfacción de clientes de este nuevo producto, estas encuestas fueron realizadas durante 1 mes.
```python
# Tu código aquí ...

import numpy as np

# Definir la media y la desviación estándar
media_capel = 4.50
desviacion_estandar_capel = 1.0  # Ajusta este valor según lo necesario

# Generar 800 respuestas de encuestas
encuestas_capel = np.random.normal(media_capel, desviacion_estandar_capel, 800)

# Asegurarse de que las respuestas estén en el rango de 1 a 7
encuestas_capel = np.clip(encuestas_capel, 1, 7)

# Definir la media y la desviación estándar para la competencia
media_competencia = 6.10
desviacion_estandar_competencia = 0.8  # Ajusta este valor según lo necesario

# Generar 500 respuestas de encuestas para la competencia
encuestas_competencia = np.random.normal(media_competencia, desviacion_estandar_competencia, 500)

# Asegurarse de que las respuestas estén en el rango de 1 a 7
encuestas_competencia = np.clip(encuestas_competencia, 1, 7)

# Definir la media y la desviación estándar para Le Pac
media_le_pac = 6.15
desviacion_estandar_le_pac = 0.8  # Ajusta este valor según lo necesario

# Generar 320 respuestas de encuestas para Le Pac
encuestas_le_pac = np.random.normal(media_le_pac, desviacion_estandar_le_pac, 320)

# Asegurarse de que las respuestas estén en el rango de 1 a 7
encuestas_le_pac = np.clip(encuestas_le_pac, 1, 7)

print("encuestas_capel = ",encuestas_capel.size)
print("encuestas_competencia = ",encuestas_competencia.size)
print("encuestas_le_pac = ",encuestas_le_pac.size)
```

## <font color='green'>**Actividad 2**</font>
### Estudiando los datasets

Para cada uno de los datasets creados se pide:

1. Graficar el correspondiente histograma y su función de masa de probabilidad y/o distribución de probabilidad según corresponda.

2. Calcular la media y desviación estandar de la muestra.

3. Calcular el porcentaje de notas 6 y 7. Este indicador en Marketing es conocido como el **Top 2 Box Score**, que corresponde a una proproción de las personas que han manifestado satisfacción por el producto/servicio.

4. Calcular el intervalo de confianza para la media y proporción Top2Box. Ambas con un nivel de significancia alpha del 0.05.




```python
import matplotlib.pyplot as plt

# Crear subgráficos en una línea
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Histograma 1: Capel
axs[0].hist(encuestas_capel, bins=10, edgecolor='black', color='blue', alpha=0.7)
axs[0].set_title('Capel')
axs[0].set_xlabel('Puntuación')
axs[0].set_ylabel('Frecuencia')

# Histograma 2: Competencia
axs[1].hist(encuestas_competencia, bins=10, edgecolor='black', color='blue', alpha=0.7)
axs[1].set_title('Competencia')
axs[1].set_xlabel('Puntuación')

# Histograma 3: Le Pac
axs[2].hist(encuestas_le_pac, bins=10, edgecolor='black', color='blue', alpha=0.7)
axs[2].set_title('Le Pac')
axs[2].set_xlabel('Puntuación')

# Ajustar el diseño para evitar solapamiento
plt.tight_layout()

# Mostrar los subgráficos
plt.show()
```

```python

from scipy.stats import norm

# Definir los datos y características de cada dataset
datasets = [
    {'nombre': 'Capel', 'encuestas': encuestas_capel, 'media': media_capel, 'desviacion_estandar': desviacion_estandar_capel},
    {'nombre': 'Competencia', 'encuestas': encuestas_competencia, 'media': media_competencia, 'desviacion_estandar': desviacion_estandar_competencia},
    {'nombre': 'Le Pac', 'encuestas': encuestas_le_pac, 'media': media_le_pac, 'desviacion_estandar': desviacion_estandar_le_pac}
]

# Configurar la disposición de subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, dataset in enumerate(datasets):
    # Generar datos para la distribución normal
    x = np.linspace(1, 7, 100)
    y = norm.pdf(x, dataset['media'], dataset['desviacion_estandar'])

    # Crear el gráfico de distribución en el subgráfico correspondiente
    axs[i].plot(x, y, color='blue', label='Distribución')
    axs[i].fill_between(x, y, color='blue', alpha=0.2)
    axs[i].axvline(dataset['media'], color='red', linestyle='dashed', linewidth=1, label=f'Media : {dataset["media"]:.2f}')
    axs[i].axvline(dataset['media'] + dataset['desviacion_estandar'], color='green', linestyle='dashed', linewidth=1, label=f'Desviación Estándar : {dataset["desviacion_estandar"]:.2f}')
    axs[i].axvline(dataset['media'] - dataset['desviacion_estandar'], color='green', linestyle='dashed', linewidth=1)
    axs[i].set_title(f'Distribución  para {dataset["nombre"]}')
    axs[i].set_xlabel('Puntuación')
    axs[i].set_ylabel('Densidad de Probabilidad')
    axs[i].legend()

# Ajustar el espacio entre subgráficos
plt.tight_layout()

# Mostrar los subgráficos
plt.show()
```

```python

from scipy import stats

# Calcular la media y la desviación estándar de la muestra para Capel
media_capel = np.mean(encuestas_capel)
desviacion_estandar_capel = np.std(encuestas_capel, ddof=1)

# Calcular el porcentaje de notas 6 y 7 (Top 2 Box Score) para Capel
top2box_score_capel = np.mean((encuestas_capel == 6) | (encuestas_capel == 7)) * 100

# Calcular el intervalo de confianza para la media de Capel
conf_interval_media_capel = stats.t.interval(0.95, len(encuestas_capel)-1, loc=media_capel, scale=stats.sem(encuestas_capel))

# Calcular el intervalo de confianza para la proporción Top 2 Box de Capel
n_top2box_capel = np.sum((encuestas_capel == 6) | (encuestas_capel == 7))
p_top2box_capel = n_top2box_capel / len(encuestas_capel)
conf_interval_proporcion_capel = stats.norm.interval(0.95, loc=p_top2box_capel, scale=np.sqrt((p_top2box_capel * (1 - p_top2box_capel)) / len(encuestas_capel)))

# Hacer lo mismo para el conjunto de datos de Competencia
media_competencia = np.mean(encuestas_competencia)
desviacion_estandar_competencia = np.std(encuestas_competencia, ddof=1)
top2box_score_competencia = np.mean((encuestas_competencia == 6) | (encuestas_competencia == 7)) * 100
conf_interval_media_competencia = stats.t.interval(0.95, len(encuestas_competencia)-1, loc=media_competencia, scale=stats.sem(encuestas_competencia))
n_top2box_competencia = np.sum((encuestas_competencia == 6) | (encuestas_competencia == 7))
p_top2box_competencia = n_top2box_competencia / len(encuestas_competencia)
conf_interval_proporcion_competencia = stats.norm.interval(0.95, loc=p_top2box_competencia, scale=np.sqrt((p_top2box_competencia * (1 - p_top2box_competencia)) / len(encuestas_competencia)))

# Hacer lo mismo para el conjunto de datos de Le Pac
media_le_pac = np.mean(encuestas_le_pac)
desviacion_estandar_le_pac = np.std(encuestas_le_pac, ddof=1)
top2box_score_le_pac = np.mean((encuestas_le_pac == 6) | (encuestas_le_pac == 7)) * 100
conf_interval_media_le_pac = stats.t.interval(0.95, len(encuestas_le_pac)-1, loc=media_le_pac, scale=stats.sem(encuestas_le_pac))
n_top2box_le_pac = np.sum((encuestas_le_pac == 6) | (encuestas_le_pac == 7))
p_top2box_le_pac = n_top2box_le_pac / len(encuestas_le_pac)
conf_interval_proporcion_le_pac = stats.norm.interval(0.95, loc=p_top2box_le_pac, scale=np.sqrt((p_top2box_le_pac * (1 - p_top2box_le_pac)) / len(encuestas_le_pac)))

# Imprimir resultados para Capel
print("Resultados para Capel:")
print(f"Media de la muestra: {media_capel:.2f}")
print(f"Desviación estándar de la muestra: {desviacion_estandar_capel:.2f}")
print(f'Top 2 Box Score: {top2box_score_capel:.2f}%')
print(f'Intervalo de Confianza para la Media: {conf_interval_media_capel}')
print(f'Intervalo de Confianza para la Proporción Top 2 Box: {conf_interval_proporcion_capel}\n')

# Imprimir resultados para Competencia
print("Resultados para Competencia:")
print(f"Media de la muestra: {media_competencia:.2f}")
print(f"Desviación estándar de la muestra: {desviacion_estandar_competencia:.2f}")
print(f'Top 2 Box Score: {top2box_score_competencia:.2f}%')
print(f'Intervalo de Confianza para la Media: {conf_interval_media_competencia}')
print(f'Intervalo de Confianza para la Proporción Top 2 Box: {conf_interval_proporcion_competencia}\n')

# Imprimir resultados para Le Pac
print("Resultados para Le Pac:")
print(f"Media de la muestra: {media_le_pac:.2f}")
print(f"Desviación estándar de la muestra: {desviacion_estandar_le_pac:.2f}")
print(f'Top 2 Box Score: {top2box_score_le_pac:.2f}%')
print(f'Intervalo de Confianza para la Media: {conf_interval_media_le_pac}')
print(f'Intervalo de Confianza para la Proporción Top 2 Box: {conf_interval_proporcion_le_pac}')
```

## <font color='green'>**Actividad 3**</font>
### Creando nuestro test de hipótesis

Desarrollar un test de hipótesis de 1 muestra para al menos 1 de los 3 datasets. Aquí Ud. puede formular la hipótesis a probar. Por ejemplo, puede ser contra un target definido por la industria o retail.



## Vamos a suponer que estamos interesados en probar si la media de satisfacción es igual a 5 (como valor hipotético).

```python
# Tu código aquí ...

from scipy.stats import ttest_1samp

# Definir la hipótesis nula: La media de satisfacción es igual a 5
hipotesis_nula_valor = 5

# Realizar el test de hipótesis para Capel
test_result_capel = ttest_1samp(encuestas_capel, hipotesis_nula_valor)

# Realizar el test de hipótesis para Competencia
test_result_competencia = ttest_1samp(encuestas_competencia, hipotesis_nula_valor)

# Realizar el test de hipótesis para Le Pac
test_result_le_pac = ttest_1samp(encuestas_le_pac, hipotesis_nula_valor)

# Imprimir resultados para Capel
print("Resultados para Capel:")
print(f'Test t = {test_result_capel.statistic:.4f}, p-value = {test_result_capel.pvalue:.4f}')

# Imprimir resultados para Competencia
print("\nResultados para Competencia:")
print(f'Test t = {test_result_competencia.statistic:.4f}, p-value = {test_result_competencia.pvalue:.4f}')

# Imprimir resultados para Le Pac
print("\nResultados para Le Pac:")
print(f'Test t = {test_result_le_pac.statistic:.4f}, p-value = {test_result_le_pac.pvalue:.4f}')
```

## <font color='green'>**Actividad 4**</font>
### Validación de algunos test de hipótesis

Desarrollar los siguientes test de hipótesis de 2 muestras para:

1. Verificar si la satisfacción de Le Pac es mayor a la del producto Capel.

2. Verificar si la satisfacción de Le Pac es mayor a la Competencia.

```python
# Tu código aquí ...

from scipy.stats import ttest_ind

# Test de hipótesis para comparar la satisfacción de Le Pac con Capel
test_result_le_pac_vs_capel = ttest_ind(encuestas_le_pac, encuestas_capel, alternative='greater')

# Test de hipótesis para comparar la satisfacción de Le Pac con Competencia
test_result_le_pac_vs_competencia = ttest_ind(encuestas_le_pac, encuestas_competencia, alternative='greater')

# Imprimir resultados
print("Test de hipótesis para comparar la satisfacción de Le Pac con Capel:")
print(f'Test t = {test_result_le_pac_vs_capel.statistic:.4f}, p-value = {test_result_le_pac_vs_capel.pvalue:.4f}')

print("\nTest de hipótesis para comparar la satisfacción de Le Pac con Competencia:")
print(f'Test t = {test_result_le_pac_vs_competencia.statistic:.4f}, p-value = {test_result_le_pac_vs_competencia.pvalue:.4f}')

# Interpretación de resultados
alpha = 0.05

if test_result_le_pac_vs_capel.pvalue < alpha:
    print("\nSe rechaza la hipótesis nula para la comparación Le Pac vs Capel. La satisfacción de Le Pac es mayor.")
else:
    print("\nNo se puede rechazar la hipótesis nula para la comparación Le Pac vs Capel. No hay evidencia suficiente para afirmar que la satisfacción de Le Pac es mayor.")

if test_result_le_pac_vs_competencia.pvalue < alpha:
    print("Se rechaza la hipótesis nula para la comparación Le Pac vs Competencia. La satisfacción de Le Pac es mayor.")
else:
    print("No se puede rechazar la hipótesis nula para la comparación Le Pac vs Competencia. No hay evidencia suficiente para afirmar que la satisfacción de Le Pac es mayor.")
```
