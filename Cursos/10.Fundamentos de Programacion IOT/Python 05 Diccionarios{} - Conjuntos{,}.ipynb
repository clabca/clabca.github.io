# Bloque A : Diccionarios { llave:valor , llave2:valor2 }

### ¿Qué es un diccionario?

• Un diccionario es una colección de elementos (par: llave, valor) que no

posee orden.

• Para acceder al valor asociado a cada elemento se utiliza la llave.

>>> alumnos = {‘Juan’: 10, ‘Pedro’: 28, ‘Ana’: 15}

>>> alumnos[‘Juan’]

10

- Las llaves deben ser únicas.

- Sólo se admiten elementos inmutables: números (int, float), strings o tuplas.

### Creación de diccionarios

>>> dicc1 = {}

>>> dicc2 = dict()

>>> dicc3 = {1: 2, 3: 4, 5: 6}

>>> dicc3[5]

6

### Agregar elementos al diccionario

>>> d = {}

>>> d['a'] = 5

>>> d['x'] = 12

>>> d['m'] = 8

>>> d

{'a': 5, 'm': 8, 'x': 12}


>>> d[‘f’] = 5

>>> d[‘x] = -10

>>> d

{‘a’: 5, ‘f’: 5, ‘m’: 8, ‘x’: -10}

### Operaciones sobre diccionarios
>>> patas = {‘humano’: 2, ‘pulpo’: 8, ‘perro’: 4, ‘gato’: 4, ‘ciempies’: 100 }

>>> list(patas)

[‘humano’, ‘pulpo’, ‘perro’, ‘gato’, ‘ciempies’]

>>> 'perro' in patas

True

>>> 8 in patas

False

>>> len(patas)

5

>>> del patas['pulpo']

>>> patas

{‘humano’: 2, ‘perro’: 4, ‘gato’: 4, ‘ciempies’: 100}

```python
artistas = {
    # nombre: codigo, origen, duracion en minutos de la presentacion)
    'Marco Antonio Solis': (1, 'Estados Unidos', 74),
    'Daddy Yankee': (2, 'Puerto Rico', 70),
    'Myriam Hernandez': (3, 'Chile', 40),
    'Los Charros de Lumaco': (4, 'Chile', 25),
    'Metallica': (5, 'Estados Unidos', 45),
    'U2': (6, 'Irlanda', 80),
    'Justin Bieber': (7, 'Canada', 5),
}

artistas_por_dia = {
    # dia, orden de las presentaciones
    "Lunes": (1, 4, 3, 6, 2, 5),
    "Martes": (2, 5, 1),
    "Miercoles": (3, 6, 2, 4),
    "Jueves": (2, 5),
}
```

```python
def cantidad_de_artistas(dia):
  cantart = 0
  cantart = len(artistas_por_dia[dia])
  return cantart

def nombre_primer_artista(dia):
  nompriart = ''
  idart = artistas_por_dia[dia][0]
  for nombre, dat in artistas.items():
    id,pais,tiempo = dat
    if idart == id:
      nompriart=nombre
  return nompriart

def nombre_primer_artista_2(dia):
  idart = artistas_por_dia[dia][0]
  for nombre in artistas:
    id = artistas[nombre][0]
    if idart == id:
      return nombre


def pais_origen_ultimo(dia):
  paisoriultart = ''
  idart = artistas_por_dia[dia][-1]
  for nombre, dat in artistas.items():
    id,pais,tiempo = dat
    if idart == id:
      paisoriultart=pais
  return paisoriultart

def pais_origen_ultimo_2(dia):
  idart = artistas_por_dia[dia][-1]
  for nombre in artistas:
    id = artistas[nombre][0]
    if idart == id:
      return artistas[nombre][1]

def tiempo_total(dia):
  tiempototal=0
  idart = artistas_por_dia[dia]
  for nombre, dat in artistas.items():
    id,pais,tiempo = dat
    if id in idart:
      tiempototal+=tiempo
  return tiempototal
 

dia = input('Ingrese dia: ')

print('Ese dia se presentaran', cantidad_de_artistas(dia), 'artistas')
print('El primer artista del dia sera', nombre_primer_artista_2(dia))
print('El ultimo artista del dia viene de', pais_origen_ultimo_2(dia))
print('Ese dia el concierto completo durara', tiempo_total(dia),'minutos')
```

### Conjuntos

Colección de valores que no tiene orden y no acepta repeticiones.

>>> colores = {‘rojo’, ‘gris’, ‘azul’, ‘gris’}

>>> colores

set([‘azul’, ‘gris’, ‘rojo’])

>>> set(‘abracadabra’)

set([‘a’, ‘b’, ‘c’, ‘d’, ‘r’])

### Agregar y eliminar cosas de un conjunto

>>> a = set()

>>> a.add(19)

>>> a.add(12)

>>> a.add(8)

>>> a.add(12)

>>> a
set([8, 12, 19])

>>> a.remove(19)
>>> a
set([8, 12])

>>> 12 in a
True

>>> len(a)
2

### Operaciones sobre conjuntos

Al utilizar esta estructura de datos podemos sacar provecho de las operaciones
matemáticas de conjuntos!

>>> a = {1, 2, 3, 4}

>>> b = {2, 4, 6, 8}


>>> a & b #interseccion

>>> set([2, 4])

>>> a | b #union

>>> set([1, 2, 3, 4, 6, 8])

>>> b – a #diferencia

>>> set([6, 8])

>>> {1, 3} < {1, 2, 3} #subconjunto

>>> True
