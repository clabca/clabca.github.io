### Patrones comunes

• Antes de escribir un programa, debe analizarse si es posible usar alguna
solución usada previamente, en un problema similar. Esto se llama “patrón de
código”.

• Por ejemplo, algunos patrones son:

1. Sumas cosas
2. Multiplicar cosas
3. Contar cosas
4. Encontrar el máximo
5. Encontrar el mínimo
6. Generar pares de cosas (combinaciones) 10

### 1. Sumar cosas

Sintaxis del patrón (pseudocódigo):

suma = 0
ciclo:

n = calcular()
suma = suma + n

Ejemplo:

Escriba una programa que reciba como entrada un número entero, y que
entregue por pantalla la suma de los números al cuadrado desde 1 hasta el
valor ingresado como parámetro.

```python
suma = 0
n = int(input('Ingrese n: '))
i = 1
while i <= n: 
	suma += i ** 2
	i += 1
print('La suma es', suma)
```

### 2. Multiplicar cosas

Sintaxis del patrón (pseudocódigo):

prod = 1
ciclo:

n = calcular()
prod = prod * n

Ejemplo: Escriba un programa que a partir de un número entregado,
calcule el factorial de dicho número.

```python
prod = 1
n = int(input('Ingrese n: '))
i = 1
while i <= n: 
	prod *= i
	i += 1
print('El factorial es', prod)
```

### 3. Contar cosas

Sintaxis del patrón (pseudocódigo):

cuenta = 0
ciclo:

...
cuenta = cuenta + 1

Ejemplo: Escriba un programa que solicite el ingreso de n números y luego
entregue la cantidad de números impares ingresados.

    Ingrese n: 4
    9
    6
    4
    7
    La cantidad de impares ingresados es: 2

```python
c = 0
n = int(input('Ingrese n: '))
i = 0
while i < n: 
	num = int(input())
	if num % 2 != 0: 
		c += 1
	i += 1
print('La cantidad de impares ingresados es:', c)
```

### Ejercicio #4

Un viajero desea saber cuánto tiempo tomó el viaje que realizó. Él tiene la duración en minutos de cada uno de los tramos del viaje.
Escriba un programa que permita ingresar los tiempos de viaje de los tramos y

El programa debe dejar de pedir tiempos cuando se ingrese un 0.

Duración del tramo: 15

Duración del tramo: 30

Duración del tramo: 87

Duración del tramo: 0

```python
continuar = True
suma = 0
while continuar:
    tiempo = int(input('Duracion tramo: '))
    if tiempo == 0:
        continuar = False
    else:
        suma = suma + tiempo
horas = suma // 60
minutos = suma % 60
print('Tiempo total de viaje:', str(horas) + ':' + str(minutos), 'horas')
```

- Si el número es impar, se debe multiplicar por 3 y sumarle 1.

Con esto se obtiene el siguiente número de la sucesión, al cual se le aplican las mismas operaciones. La sucesión termina cuando el número obtenido es 1.

### 4. Encontrar el máximo

Sintaxis del patrón:

mayor = numero_muy_chico

ciclo:

n = calcular()

if n > mayor:

mayor = n

Ejemplo: Escriba un programa que solicite al usuario la cantidad de
números a ingresar, luego solicite los números y, al finalizar, muestre el
número mayor ingresado.

    Ingrese cantidad de números: 5
    4

    7
    5

```python
mayor = -float('inf')
n = int(input('Ingrese cantidad de numeros: '))
i = 0
while i < n: 
	num = int(input())
	if num > mayor: 
		mayor = num
	i += 1
print('El mayor es', mayor)
```

### Ejercicio #1

Escriba un programa que permita determinar el número mayor de n
números solicitados al usuario.

    Ingrese n: 4
    Ingrese numero 1: 99
    Ingrese numero 2: -3
    Ingrese numero 3: 83
    Ingrese numero 4: 198
    El mayor es el 198 que corresponde al numero 4

```python
n = int(input('Ingrese n: '))
i = 0
posicion = 0
mayor = -float('inf')
while i < n:
    numero = int(input('Ingrese numero ' + str(i + 1) + ': '))
    if numero > mayor:
        mayor = numero
        posicion = i
    i += 1
print('El mayor es', mayor, 'que corresponde al numero', posicion + 1)
```

### Ejercicio #2

En estadística descriptiva, se define el rango de un conjunto de datos reales
como la diferencia entre el mayor y el menor de los datos.

Por ejemplo, si los datos son [5.96, 6.74, 7.43, 4.99, 7.20, 0.56,
2.80] entonces el rango será: 7.43 – 0.56 = 6.87

Escriba un programa que; (1) pregunte al usuario cuántos datos ingresará, (2)
pida los datos uno por uno y (3) entregue como resultado el rango de los
datos (suponga que todos los datos que ingrese serán válidos).

```python
n = int(input('Cuantos datos ingresara?: '))
i = 0
menor = float('inf')
mayor = -float('inf')
while i < n: 
	num = float(input('Ingrese num: '))
	if num < menor: 
		menor = num
	if num > mayor: 
		mayor = num
	i += 1 
print('El rango es', round(mayor - menor, 2))
```

### Ejercicio #4

Un jugador debe lanzar dos dados numerados del 1 al 6, y su puntaje es la suma de los valores obtenidos.

Un puntaje dado, puede ser obtenido con varias combinaciones posibles. Por
ejemplo, el puntaje 4 se logra con 3 combinaciones;1+3, 2+2, 3+1


Escriba un programa que pregunte al usuario un puntaje, y muestre como resultado
la cantidad de combinaciones con las que se puede obtener dicho puntaje.


Ingrese el puntaje: 4
Hay 3 combinaciones para obtener 4

```python
puntaje = int(input('Ingrese puntaje: '))
d1 = 1
combinaciones = 0
while d1 <= 6:
    d2 = 1
    while d2 <= 6:
        if d1 + d2 == puntaje:
            combinaciones += 1
        d2 += 1
    d1 += 1
print('Hay', combinaciones, 'combinaciones para obtener', puntaje)
```

Escriba un programa que ingrese la cantidad total de benefactores N y el monto total a recaudar M.

Su programa debe solicitar el nombre y el monto que donará cada benefactor, uno por uno. Su programa debe detenerse una vez que se alcance el monto total M, o cuando se terminen de ingresar los N benefactores.


Indique cuál es el monto mayor donado y qué benefactor lo donó.

```python
benefactores = int(input("ingrese numero maximo de benefactores : "))
monto = int(input("Ingrese monto maximo a recaudar : "))
ben=0
suma=0
maxdonador =""
maxmonto = -float('inf')
while suma<monto and benefactores>ben:
  nombrebenefactor = input("Ingrese Nombre del Benefactor : ")
  montobenefactor = int(input("Ingrese monto donado : "))
  if montobenefactor>maxmonto:
    maxdonador = nombrebenefactor
    maxmonto = montobenefactor
  suma=suma+montobenefactor
  ben = ben + 1
print("Benefactores : ",ben)
print("Monto Reudado : ",suma)
print("Mejor Benefactor : ",maxdonador)
print("Monto Reudado : ",maxmonto)
```

# Tarea 2

El banco CBI necesita implementar mejoras en sus mecanismos de seguridad, precisamente en los que están relacionados con la generación de claves iniciales para sus clientes de tarjetas de crédito. Para esto, desarrolle un programa que permita generar una clave de 4 dígitos, usando el siguiente algoritmo: 

• Leer tantos números como sea necesario para generar la clave. 

• Para cada número deberá recorrer su dígitos y encontrar el menor.

• Si el dígito menor es par, pasará a formar parte de la clave, en caso contrario, se procesa el siguiente número hasta formar la clave de 4 dígitos.

• Mostrar la clave obtenida.

```python

p1 = 0
p2 = 0
p3 = 0
p4 = 0
while p1==0 or p2==0 or p3==0 or p4==0:
  numero = int(input('Ingrese numero: '))
  menor = float('inf')
  while numero>0:
    d1 = numero % 10
    numero = numero // 10
    if d1 < menor:
      menor = d1
  if menor%2 == 0:
      if p1==0:
        p1 = menor
        print("primer digito" ,menor)
      elif p2==0:
        p2 = menor
        print("Segundo digito" ,menor)
      elif p3==0:
        p3 = menor
        print("tercer digito" ,menor)
      else:
        p4 = menor
        print("Cuarto digito" ,menor)

print('La Clave obtenida es',p1,p2,p3,p4)
```
