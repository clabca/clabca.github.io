### Trabajo grupal – Ejercicio #1

Escriba un programa que abra el archivo quijote.txt y cuente:

- Cuántas letras tiene

- Cuántas palabras tiene

- Cuántas líneas tiene

```python
archivo = open('/content/drive/My Drive/Colab Notebooks/quijote.txt')
letras = 0
palabras = 0
lineas = 0
for linea in archivo:
  linea = linea.strip()
  letras += len(linea.replace(' ',''))
  palabras += len(linea.split())
  lineas +=1
print('Tiene', letras ,' Letras')
print('Tiene', palabras ,' Palabras')
print('Tiene', lineas ,' Lineas')

archivo.close()
```

temp = {

Escriba la función crear_reporte(temperaturas),

cuyo parámetro es el diccionarios de temperaturas, y que genere el archivo de

Ejemplo archivo generado: reporte.txt

```python
temp = {
'Vina del mar': (9, 26),
'Valparaiso': (10, 24),
'Quilpue': (7, 30),
'Quintero': (4, 22)
}



def crear_reporte(temperaturas):
  archivo = open('/content/drive/My Drive/Colab Notebooks/reporte.txt', 'w')
  for ciudad, t in temperaturas.items():
    linea = ciudad + ':max '+ str(t[1])+'   min '+str(t[0])+'\n'
    archivo.write(linea)
  archivo.close

crear_reporte(temp)
```

### Método útil de strings: split

Este método permite cortar un string en elementos definidos por el separador que recibe como parámetro. El resultado siempre será una lista de strings con el resultado de la
separación.

>>> cadena = ‘Juan:Perez:15753829-4:Chillan Viejo’


>>> cadena.split(‘:’)

[‘Juan’, ‘Perez’, ‘15753829-4’, ‘Chillan Viejo’]


En caso que split no reciba parámetros, Python interpreta que el separador es un espacio.


>>> cadena.split()

[‘Juan:Perez:15753829-4:Chillan’, ‘Viejo’]

```python
cadena = 'Juan:Perez:15753829-4:Chillan Viejo'
cadena.split(':')
```

### Método útil para listas de strings: join

El método join() permite unir los elementos de una lista de strings, entregando un string resultante con todos los elementos unidos por un separador.

>>> lista = [‘Juan’, ‘Perez’, ‘100’]


>>> ‘;’.join(lista)

‘Juan;Perez;100’

>>> ‘X’.join(lista)

‘JuanXPerezX100’


```python
lista = ['Juan', 'Perez', '100']
';'.join(lista)
```

```python
'X'.join(lista)
```

### Función útil: map

La función map permite mapear una función en un elemento iterable. Así, no
será necesario recorrerlo para cambiar los elementos.

>>> list(map(int, [’10’, ‘20’, ’30’, ‘40’]))

[10, 20, 30, 40]

>>> list(map(str, [40, 50, 60, 70]))

[‘40’, ‘50’, ‘60’, ‘70’]

>>> list(map(str, (1, 2, 3, 10)))

(‘1’, ‘2’, ‘3’, ‘10’)

```python
list(map(int, ['10', '20', '30', '40']))
```

```python
list(map(str, [40, 50, 60, 70]))
```

```python
list(map(str, (1, 2, 3, 10)))
```

¿Cómo podría armar la línea usando el método join?

```python
alumnos = [
('Esteban', 'Gutierrez', [10, 20, 30]),
('Juan', 'Perez', [20, 30, 40])
]

archivo = open('/content/drive/My Drive/Colab Notebooks/alumnos_nuevo.txt', 'w')
for nombre, apellido, notas in alumnos:
  promedio = sum(notas) / len(notas)
  linea = '{};{};{}\n'.format(nombre, apellido, promedio)
  archivo.write(linea)
archivo.close()
```

### Trabajo grupal – Ejercicio #1

Las notas de un ramo están almacenadas en el archivo “notas.txt”. Cada línea tiene el nombre del alumno y sus seis notas, separadas por dos puntos (“:”)

Escriba un programa que cree un nuevo archivo llamado “reporte.txt”, en que cada línea indique si el alumno está aprobado (promedio >= 4.0) o reprobado (promedio < 4.0)


Resultado a obtener en archivo reporte.txt


Pepito aprobado

Yayita aprobado

Fulanita aprobado

Moya reprobado

```python
archivo = open('/content/drive/My Drive/Colab Notebooks/notas.txt')
archivo2 = open('/content/drive/My Drive/Colab Notebooks/reporte_notas.txt','w')
for linea in archivo:
  linea = linea.strip().split(':')
  notas = list(map(float, linea[1:]))
  promedio = sum(notas)/len(notas)
  if promedio<4:
      escribe = 'El Alumno(a) '+linea[0]+ ' esta Reprobado \n'
  else:
      escribe = 'El Alumno(a) ' +linea[0]+ ' esta Aprobado \n'
  archivo2.write(escribe)

archivo.close()
archivo2.close()
```
