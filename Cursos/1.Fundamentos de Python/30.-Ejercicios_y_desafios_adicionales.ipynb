### Obtener el producto de todos los demás elementos   

Desafío:

Dado un arreglo de enteros, calcule y retorne un nuevo arreglo, tal que, cada elemento en el indice $i$ del nuevo arreglo corresponde al producto de todos los números en el arreglo original exepto el que está en la posición $i$.

Por ejemplo:

Si el input es
``[1, 2, 3, 4, 5]``, el resultado esperado debería ser ``[120, 60, 40, 30, 24]``

``[3, 2, 1]``, el resultado esperado debería ser ``[2, 3, 6]``

Restricción:

No puedes usar divisiones.

```python
# Tú código aquí ...

def productos(nums):
    # Genera productos de números anteriores.
    producto_prefijos = []
    for num in nums:
        if producto_prefijos:
            producto_prefijos.append(producto_prefijos[-1] * num)
        else:
            producto_prefijos.append(num)
    print(producto_prefijos)

    # Genera productos de números posteriores
    producto_sufijos = []
    for num in reversed(nums):
        if producto_sufijos:
            producto_sufijos.append(producto_sufijos[-1] * num)
        else:
            producto_sufijos.append(num)
    producto_sufijos = list(reversed(producto_sufijos))
    print(producto_sufijos)

    # Genera salida con el resultado del producto de prefijos and sufijos.
    resultado = []
    for i in range(len(nums)):
        if i == 0:
            resultado.append(producto_sufijos[i + 1])
        elif i == len(nums) - 1:
            resultado.append(producto_prefijos[i - 1])
        else:
            resultado.append(producto_prefijos[i - 1] * producto_sufijos[i + 1])
    return resultado
```

```python
input = [1, 2, 3]
productos(input)
```

```python
input = [1, 2, 3, 4, 5]
productos(input)
```

### Cuenta los elementos más pequeños que están a la derecha


Desafío:

Dado un arreglo de enteros, calcule y retorne un nuevo arreglo, tal que, cada elemento del nuevo arreglo corresponde al número (conteo) de eleemntos más pequeños a la derecha de ese elemento en el arreglo original.

Por ejemplo:

Si el input es
``[3, 4, 9, 6, 1]``, el resultado esperado debería ser ``[1, 1, 2, 1, 0]``

estos es
* hay 1 elemento más pequeño a la derecha de 3
* hay 1 elemento más pequeño a la derecha de 4
* hay 2 elemento más pequeño a la derecha de 9
* hay 1 elemento más pequeño a la derecha de 6
* hay 1 elemento más pequeño a la derecha de 1

Restricción:

Desarrolla una solución que no sea $O(n)$ o superior en complejidad algorítmica.

Tip:

Explora el uso de la librería `bisect`.

```python
# Tú código aquí ...

def cuenta_pequenos(lista):
    resultado =[]
    for i, num in enumerate(lista):
        cuenta = sum(val < num for val in lista[i + 1:])
        resultado.append(cuenta)
    return resultado
```

```python
l = [3, 4, 9, 6, 1]
cuenta_pequenos(l)
```

## <font color='green'> Actividad 3
### Decorador de tiempo
Escribe un decorador que se llame __timer_decorator__, el cual mida e imprima el tiempo de ejecucuón de una función.



```python
# Tú código aquí ...

import time
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"La función {func.__name__} se ejecutó en {end_time - start_time} segundos.")
        return result
    return wrapper
```

```python
@timer_decorator
def espera(n):
    time.sleep(n)
    return n

espera(3)
```

### Clase cuenta bancaria
Defina una clase __CuentaBancaria__ que represente una cuenta bancaria. La clase debe tener un atributo __balance__ que se inicializa a 0, y los métodos __deposito__ y __retiro__ para añadir y retirar dinero de la cuenta, respectivamente. Además, __retiro__ no debe permitir que el balance se vuelva negativo, y en caso de que se intente retirar más dinero del que hay en la cuenta, debe levantar una excepción.

Pruebe la clase creando una instancia de la misma y ejecutando transacciones de depósitos y retiros.

```python
# Tú código aquí ...

class CuentaBancaria:
    def __init__(self):
        self.balance = 0

    def deposito(self, monto_dep):
        self.balance += monto_dep

    def retiro(self, monto_ret):
        if monto_ret > self.balance:
            raise ValueError("Fondos insuficientes ...")
        self.balance -= monto_ret
```

```python
# Prueba
cuenta = CuentaBancaria()
cuenta.deposito(100)
print(f'Balaance: {cuenta.balance}')
cuenta.retiro(50)
print(f'Balaance: {cuenta.balance}')
try:
    cuenta.retiro(200)
except ValueError as e:
    print(e)
```

## <font color='green'> Actividad 6
### Diagonal de matriz

Define una función __diagonal__ que tome una lista de listas representando una matriz cuadrada (es decir, una matriz con el mismo número de filas y columnas), y retorne una nueva lista que contenga los elementos en la diagonal de la matriz (es decir, los elementos en las posiciones (0, 0), (1, 1), etc.).

```python
# Tu código aquí ...

def diagonal(matrix):
    return [matrix[i][i] for i in range(len(matrix))]
```

```python
# Prueba

A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(diagonal(A)) # -> [1, 5, 9]
```

### Clase stack

Define una clase __Stack__ que implemente una pila de elementos. La clase debe tener métodos __push__ para añadir un elemento a la pila, __pop__ para remover el último elemento añadido, __peek__ para ver el último elemento añadido sin removerlo, e __info__ para imprimir la pila completa. Además, si se intenta __pop__ o __peek__ cuando la pila está vacía, debe levantar una excepción.

```python
# Tu código aquí ...

class Stack:
    def __init__(self):
        self.stack = []

    def push(self, element):
        self.stack.append(element)

    def pop(self):
        if not self.stack:
            raise IndexError("Pop from empty stack")
        return self.stack.pop()

    def peek(self):
        if not self.stack:
            raise IndexError("Peek into empty stack")
        return self.stack[-1]

    def info(self):
        return print(self.stack)
```

```python
# Prueba
s = Stack()
s.info()
s.push(1)
s.push(2)
s.info()
print(s.peek())
print(s.pop())
try:
    s.pop()
    s.pop()
except IndexError as e:
    print(e)
```

### Cálculo del cuadrado con memoria

Define una función __cuadrado__ que calcule el cuadrado de un número. Luego, defina un decorador __memoriza__ que guarde en un diccionario los resultados de las llamadas a __cuadrado__, de manera que si se vuelve a llamar a __cuadrado__ con el mismo número, retorne el resultado guardado en lugar de recalcularlo.

```python
# Tu código aquí ...

def memoriza(func):
    cache = {}
    def wrapper(n):
        if n in cache:
            print('Usando memoria')
        else:
            cache[n] = func(n)
        return cache[n]
    return wrapper

@memoriza
def cuadrado(n):
    return n * n
```

```python
# Prueba
print(cuadrado(3))
print(cuadrado(15))
print(cuadrado(3))  # Esta llamada usa el valor guardado en lugar de recalcular el factorial
```
