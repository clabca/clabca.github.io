el cual permitir que los widgets interacten con Jupyter Notebook.

Comience como de costumbre importando y configurando el entorno de `Matplotlib`, incluido el mdulo `ipywidget`, el mtodo `interact`.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from ipywidgets import interact
import ipywidgets as widgets
```

Nos creamos una funcin de ploteo para probar la interactividad.

```python
# Nos creamos una funcin de ploteo
def f(x):
    plt.plot(np.arange(-10,10), x*np.arange(-10,10))
    plt.ylim(-30,30)

f(10)
```

Al usar el mtodo `interact` y establecer `x=1` de forma predeterminada, obtenemos los siguientes puntos:

```python
def f(x):
    plt.plot(np.arange(-10, 10), x * np.arange(-10, 10))
    plt.ylim(-30, 30)
    plt.show()

interact(f, x=widgets.FloatSlider(min=-10, max=10, step=0.1, value=1))
plt.show()
```

```python
# Nos creamos una funcin de ploteo
def f(x):
    plt.plot(np.arange(-10,10), x*np.arange(-10,10))
    plt.ylim(-30,30)
    plt.show()

interact(f, x=1)
plt.show()
```

Controlamos el rango y el tamao del paso de `x`.

```python
# Controlemos el rango de x
def f(x):
    plt.plot(np.arange(-10,10), x*np.arange(-10,10))
    plt.ylim(-30,30)
    plt.show()
interact(f, x=(-2,4,0.5))
plt.show() # Mejoramos la visualizacin
```

```python
# Probemos otra funcin
d = np.arange(0, 5, 0.5)
def curve(x):
    plt.plot(d, d**x, color='r')
    plt.xlim(0,6)
    plt.ylim(-1, 20)
    plt.show()

interact(curve, x=(0,5,0.5))
plt.show()
```

Tambin podemos elegir mltiples widgets con el mtodo de interaccin. Entonces, si tenemos una funcin que tiene varios valores, como a = 1, b = 3, automticamente obtenemos un par de widgets:

```python
# Multiples widgets
def f(a,b):
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b))
    plt.title("Power Law: $x=ay^b$")
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=3)
plt.show()
```

```python
# Multiples widgets
def f(a,b):
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b))
    plt.title(f"Power Law: $x={a}y^{b}$") # Con f-strings
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=3)
plt.show()
```

```python
# Aadimos un menu dropdown
colores=['red', 'green', 'blue']

def f(a,b, color):
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b),
             color=color)
    plt.title(f"Power Law: $x={a}y^{b}$") # Con f-strings
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=3, color=colores)
plt.show()
```

```python
# Las potencias las podemos hacer discretas con un diccionario
potencias = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4}

colores=['red', 'green', 'blue']
def f(a,b, color):
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b),
             color=color)
    plt.title(f"Power Law: $x={a}y^{b}$") # Con f-strings
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=potencias, color=colores)
plt.show()
```

```python
# Ya que estamos, traduzcamos los colores
colores = {'Rojo': 'r', 'Verde':'g', 'Azul':'b'}
potencias = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4}

def f(a,b, color):
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b),
             color=color)
    plt.title(f"Power Law: $x={a}y^{b}$") # Con f-strings
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=potencias, color=colores)
plt.show()
```

```python
# Y modifiquemos el tamao del canvas (figure)
colores = {'Rojo': 'r', 'Verde': 'g', 'Azul': 'b'}
potencias = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4}
size = {'Pequeo': (4,3), 'Mediano': (6,4), 'Grande': (10,6)}

def f(a,b, color,size=(4,3)):
    plt.figure(figsize=size)
    plt.plot(np.arange(0, 10, 0.2), a*np.power(np.arange(0, 10, 0.2), b),
             color=color)
    plt.title(f"Power Law: $x={a}y^{b}$") # Con f-strings
    plt.ylim(0, 100)
    plt.show()

interact(f, a=1, b=potencias, color=colores, size=size)
plt.show()
```
